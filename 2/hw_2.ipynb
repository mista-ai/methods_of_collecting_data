```python
#from pprint import pprint
from lxml import html
import requests
import json
```

```python
url = 'https://lenta.ru/'
headers = {'User-Agent': 
           'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}

response = requests.get(url, headers = headers)
```

```python
response.url
```

```python
dom = html.fromstring(response.text)
```

```python
#card-mini
#card-big
#card-feature
#//div[contains(@class, 'card-big') or contains(@class, 'card-mini') or contains(@class, 'card-feature')]
news = dom.xpath("//a[contains(@class, 'card-big') or contains(@class, 'card-mini') or contains(@class, 'card-feature')]")
```

```python
news_dict = dict()

for piece in news:
    piece_text = piece.xpath(".//h3/text()")
    if len(piece_text) == 0:
        piece_text = piece.xpath(".//span/text()")
    piece_text = piece_text[0]
    try:
        piece_time = piece.xpath(".//time/text()")[0]
    except:
        piece_time = 'None'
    piece_link = piece.xpath(".//@href")[0]
    news_dict[piece_text] = {'link': piece_link, 'time': piece_time}
    
for k, v in news_dict.items():
    print(k, v)
```

```python
with open("lenta_news.json", "w", encoding='utf-16') as out:
    out.write(json.dumps(news_dict, ensure_ascii=False))
```
