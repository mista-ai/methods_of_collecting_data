```python
from pymongo import MongoClient
from pprint import pprint
```

```python
client = MongoClient()
```

```python
client.list_database_names()
```

```python
db = client.jobs_parser
```

```python
for job in db.hh_ru.find():
    pprint(job)
```

```python
# Вакансии с Python
i = 0
for job in db.hh_ru.find({'key_skills': 'Python'}):
    i += 1
    print(f"{i}) {job['name']}, Зарплата от {job['salary_from']} до {job['salary_to']}, {job['company']}")
```

```python
# Вакансии с зарплатой выше 100 000
i = 0
for job in db.hh_ru.find({'salary_from': {'$gt': 100000}}):
    i += 1
    print(f"{i}) {job['name']} {job['salary_from']}")
```
