1. Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.

```python
import requests
import json
from pprint import pprint
```

```python
username = 'mista-ai'
url = 'https://api.github.com/users/mista-ai/repos'
```

```python
response = requests.get(url)
```

```python
response
```

```python
repos = response.json()
```

```python
print(repos)
```

```python
for repo in repos:
    print(f"Repo name: {repo['name']}\nRepo URL: {repo['html_url']}")
```

#### Запись в файл

```python
outjson = json.dumps(response.text)
```

```python
with open("github_repo_list.json", "w") as out:
    out.write(json.dumps(repos))
```

```python
version = 5.131
user_id = 262534613
```

```python
with open('keys.txt', 'r') as f:
    access_token = f.read()
```

```python
url = "https://api.vk.com/method/groups.get"
```

```python
response = requests.get(url, 
                        params={
                            'access_token': access_token,
                            'v': version,
                            'user_id': user_id
                        })
```

```python
groups = response.json()['response']['items']
```

```python
count = response.json()['response']['count']
```

```python
url = "https://api.vk.com/method/groups.getById"
```

```python
response = requests.get(url, 
                        params={
                            'access_token': access_token,
                            'v': version,
                            'group_ids': groups
                        })
```

```python
groups = response.json()['response']
```

```python
i = 0
for group in groups:
    i = i + 1
    print(group['name'])
```
